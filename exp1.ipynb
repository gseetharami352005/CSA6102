```python
import os
import time
from datetime import datetime

def get_file_timestamps(filepath):
    """Returns creation, modification, and access timestamps of a file."""
    try:
        stat_info = os.stat(filepath)
        # Note: 'st_birthtime' (creation time) might not be available on all systems (e.g., some Linux file systems).
        # In such cases, 'st_ctime' (last metadata change time) is often reported as creation time.
        creation_time = datetime.fromtimestamp(stat_info.st_birthtime if hasattr(stat_info, 'st_birthtime') else stat_info.st_ctime)
        modification_time = datetime.fromtimestamp(stat_info.st_mtime)
        access_time = datetime.fromtimestamp(stat_info.st_atime)
        return {
            'Created': creation_time,
            'Modified': modification_time,
            'Accessed': access_time
        }
    except FileNotFoundError:
        return None

def print_timestamps(description, timestamps):
    """Prints file timestamps with a description."""
    print(f"\n--- {description} ---")
    if timestamps:
        for key, value in timestamps.items():
            print(f"{key}: {value}")
    else:
        print("File not found or timestamps could not be retrieved.")


# Define a filename for demonstration
filename = 'locards_trace_demo.txt'

# --- Step 1: Initial Creation ---
print("Demonstrating Locard's Exchange Principle with file timestamps")
print("-----------------------------------------------------------")

# Create a new file
with open(filename, 'w') as f:
    f.write("Initial content.\n")

print(f"1. Created file: '{filename}'")
initial_timestamps = get_file_timestamps(filename)
print_timestamps("After Creation", initial_timestamps)

# --- Step 2: Modification ---
# Wait a short moment to ensure distinct timestamps (system clock granularity)
time.sleep(2)

# Modify the file
with open(filename, 'a') as f:
    f.write("Content added during modification.\n")

print(f"\n2. Modified file: '{filename}'")
modified_timestamps = get_file_timestamps(filename)
print_timestamps("After Modification", modified_timestamps)

# --- Step 3: Access (Read) ---
# Wait a short moment
time.sleep(2)

# Access the file (read its content)
with open(filename, 'r') as f:
    content = f.read()

print(f"\n3. Accessed file (read content): '{filename}'")
# print(f"Content read: \n{content}") # Uncomment to see content
accessed_timestamps = get_file_timestamps(filename)
print_timestamps("After Access (Read)", accessed_timestamps)

# --- Cleanup ---
# Wait a short moment before deleting
time.sleep(1)

os.remove(filename)
print(f"\n4. Cleaned up: Deleted '{filename}'")
print("\nObservation: Notice how the 'Modified' and 'Accessed' timestamps change after each interaction, leaving a digital trace.")
```
