```python
import os
import time
from datetime import datetime

# File name
filename = "evidence_demo.txt"

# Step 1: Create the file
with open(filename, "w") as file:
    file.write("Original evidence.\n")

print("File created.")

# Function to display file metadata
def show_metadata(file):
    stats = os.stat(file)
    print("\nFile Metadata")
    print("-" * 30)
    print("File Name :", file)
    print("Created   :", datetime.fromtimestamp(stats.st_ctime))
    print("Modified  :", datetime.fromtimestamp(stats.st_mtime))
    print("Accessed  :", datetime.fromtimestamp(stats.st_atime))

# Display original metadata
show_metadata(filename)

# Wait a few seconds
time.sleep(5)

# Step 2: Modify the file
with open(filename, "a") as file:
    file.write("Additional evidence added.\n")

print("\nFile modified.")

# Display updated metadata
show_metadata(filename)
```
