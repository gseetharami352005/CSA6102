```python
import hashlib
import os

def calculate_file_hash(filepath, hash_algorithm='sha256'):
    """Calculates the hash of a file using the specified algorithm."""
    hasher = hashlib.new(hash_algorithm)
    try:
        with open(filepath, 'rb') as f:
            while chunk := f.read(4096):
                hasher.update(chunk)
        return hasher.hexdigest()
    except FileNotFoundError:
        return None


# Define a filename for demonstration
filename = 'digital_evidence.txt'

print("Demonstrating Chain of Custody Verification with File Hashing")
print("-----------------------------------------------------------\n")

# --- Step 1: Create Initial Digital Evidence ---
initial_content = "This is the original digital evidence. It should not be altered."
with open(filename, 'w') as f:
    f.write(initial_content)

print(f"1. Created initial evidence file: '{filename}'")
initial_hash = calculate_file_hash(filename)
print(f"   Initial SHA256 Hash: {initial_hash}\n")

# --- Step 2: Simulate an unauthorized modification ---
# This change breaks the chain of custody
modified_content = "This is the original digital evidence. It should not be altered. -- SOMEONE TAMPERED HERE!"
with open(filename, 'w') as f:
    f.write(modified_content)

print(f"2. Modified evidence file: '{filename}' (simulating tampering)")
modified_hash = calculate_file_hash(filename)
print(f"   Modified SHA256 Hash: {modified_hash}\n")

# --- Step 3: Verify the Chain of Custody ---
print("3. Verifying Chain of Custody...")
if initial_hash == modified_hash:
    print("   Verification SUCCESS: The file hash has NOT changed. Chain of custody is intact.")
else:
    print("   Verification FAILED: The file hash HAS changed. Chain of custody is BROKEN.")
    print("   (This indicates the digital evidence has been altered.)\n")

# --- Cleanup ---
os.remove(filename)
print(f"4. Cleaned up: Deleted '{filename}'")
```
