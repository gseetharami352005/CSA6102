```python
# Digital Forensic Investigation Report Generator

# Collect case details
case_id = input("Enter Case ID: ")
investigator = input("Enter Investigator Name: ")
date = input("Enter Investigation Date (DD-MM-YYYY): ")
device = input("Enter Device Examined: ")

# Generate report
print("\n" + "=" * 60)
print("      DIGITAL FORENSIC INVESTIGATION REPORT")
print("=" * 60)

print(f"\nCase ID        : {case_id}")
print(f"Investigator   : {investigator}")
print(f"Date           : {date}")
print(f"Device Examined: {device}")

print("\n1. Identification")
print("-" * 60)
print("• Identified the digital device and potential evidence.")
print("• Secured the crime scene and documented the evidence.")

print("\n2. Preservation")
print("-" * 60)
print("• Preserved the evidence using forensic procedures.")
print("• Maintained chain of custody to ensure integrity.")

print("\n3. Collection")
print("-" * 60)
print("• Collected forensic images of the storage media.")
print("• Stored evidence in a secure forensic repository.")

print("\n4. Examination and Analysis")
print("-" * 60)
print("• Examined files, logs, metadata, and timestamps.")
print("• Analyzed evidence for suspicious activities.")
print("• Verified evidence integrity using cryptographic hashes.")

print("\n5. Reporting")
print("-" * 60)
print("• Documented findings and conclusions.")
print("• Prepared the forensic report for legal proceedings.")

print("\nConclusion")
print("-" * 60)
print("The investigation followed the standard five-stage")
print("digital forensic investigation lifecycle while")
print("maintaining evidence integrity and chain of custody.")

print("\n" + "=" * 60)
print("End of Report")
print("=" * 60)
```
